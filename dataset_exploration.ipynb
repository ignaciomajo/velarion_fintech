## <font color='#315aa0'>**Mapa a seguir**</font>

### 1. Preparación de clientes (df_churn)

* **Mantener todas las columnas originales de df_churn, especialmente:**

    - CustomerId (ID único)
    - Exited (target)
    - Balance, Tenure, CreditScore, Age, NumOfProducts, etc.
    - Generar columna “Mes de abandono” para clientes que hicieron churn (Exited=1):
        - Aleatorio entre 0 y 6 meses antes del final del período simulado.
        - Esto definirá hasta cuándo tendrán transacciones.
    - Crear columna “Balance inicial” que será el punto de partida para generar oldbalanceOrig y newbalanceOrig en transacciones.

### 2. Preparación de transacciones (df_paysim)

* **Filtrar y seleccionar columnas relevantes:**

    - step → será transformado en fecha
    - type → tipo de transacción
    - amount → monto, escalado y limitado por balance
    - nameOrig → será reemplazado por CustomerId
    - Opcional: oldbalanceOrg, newbalanceOrig → recalculados para coherencia
    - isFraud → puede generar feature de “fraudulent activity”

* **Escalar montos:**

    - `amount_scaled = (amount_original / max_amount_paysim) * balance_cliente`


Esto asegura que ninguna transacción supere el balance del cliente.

* **Convertir step en fechas:**

    - Definir fecha inicial ficticia, p.ej. "2023-01-01".
    - Transformar step en días o meses:

        - `fecha_transaccion = fecha_inicio + timedelta(days=step)`

Permite agregar features temporales: actividad por mes, últimos 6 meses, etc.

### 3. Asignación de transacciones a clientes

* Reemplazar nameOrig por CustomerId de df_churn.

* **Reglas de asignación:**

    - Clientes que no hicieron churn (Exited=0) → asignar transacciones a lo largo de todo el período simulado.
    - Clientes que hicieron churn (Exited=1) → asignar transacciones hasta “mes de abandono”.

* **Opciones de distribución:**
    - Aleatoria, pero proporcional al balance o al tenure.
    - Para que haya variedad en la cantidad de transacciones por cliente.

### 4. Generación de balances coherentes

**Para cada cliente:**

oldbalanceOrg = balance_anterior
newbalanceOrig = balance_anterior - amount_scaled (para retiros) o + amount_scaled (para depósitos)

Esto mantiene consistencia interna en la historia de transacciones.

### 5. Agregación y creación de features por cliente

Estas serán las columnas que usaría el modelo de ML:

**Activity Features:**

* Total de transacciones
* Promedio de monto por transacción
* Número de transacciones por tipo (CASH_OUT, PAYMENT, etc.)
* Último mes de actividad
* Varianza de montos

**Balance Features:**

* Promedio de saldo (newbalanceOrig)
* Máximo saldo
* Último saldo registrado

**Fraud / Riesgo:**

* Número de transacciones fraudulentas (isFraud)
* Proporción de transacciones grandes (> X% del saldo)

**Temporal:**

* Tendencia de actividad (disminución antes del churn)
* Mes desde la última transacción

### 6. Dataset final para ML

**Filas:** un cliente por fila

**Columnas:**

* Todas las columnas originales de df_churn (información estática)
* Features agregadas de transacciones (actividad, balances, fraude, temporal)

**Target:** Exited

# <font color='#d3d1d1'>**Manipulación del dataset**</font>

## 🔑 Principios para evitar “dataset basura”

### 1. Definí hipótesis de negocio realistas
Antes de tocar los datos, escribí supuestos que tengan sentido en un banco real:

- Clientes con menos interacciones en la app → más probabilidad de churn.
- Disminución en frecuencia de transacciones en los últimos meses → mayor riesgo de churn.
- Edad y salario influyen en la estabilidad del cliente (jóvenes y bajos salarios = más churn).
- Factores externos como inflación alta → más churn en clientes de bajo ingreso.

> 👉 Esto asegura que los patrones que generes no sean aleatorios, sino consistentes con lo que podría pasar en la realidad.

---

### 2. Usá correlaciones en lugar de ruido puro
En lugar de asignar variables externas o de interacción al azar:

- Haz que estén correlacionadas con churn de forma controlada.

**Ejemplo:**
- Clientes que hicieron churn tienen un 40% menos de interacciones con la app en los últimos 3 meses.
- Clientes de churn tienen un 20% más de transacciones fallidas o rechazadas.
- Factores externos (ej: inflación simulada) afectan más a clientes con Balance < X.

> 👉 Esto genera señales diferenciales para el modelo.

---

### 3. Introduce variabilidad dentro de cada grupo
No todos los clientes que hicieron churn deben tener baja actividad. Introducí distribuciones:

- 60% de churners tienen muy baja interacción.
- 30% tienen actividad media.
- 10% siguen siendo activos hasta el final.

> 👉 Esto evita que el modelo aprenda reglas triviales (del tipo: “si transacciones = 0 → churn”).

---

### 4. Usá ruido controlado (stochastic noise)
Un poco de azar es necesario para que no todo sea determinístico, pero debe ser pequeño en comparación con la señal principal.

**Ejemplo:**  
Número de interacciones promedio de un churner = 20, con desviación estándar de ±5.

---

### 5. Validación manual antes del modelo
Antes de entrenar, explorá el dataset final con gráficos:

- Histograma de transacciones churn vs no churn.
- Interacciones app churn vs no churn.
- Factores externos según balance.

> 👉 Si ves diferencias razonables entre grupos, entonces tu dataset tiene patrones útiles.

---

## 🔧 Concretando para tus 4 datasets

### 1. Churn (base)
- Target + atributos estáticos.

### 2. Transacciones
- Clientes que hicieron churn deben mostrar una caída de actividad en los últimos meses.
- Clientes que no hicieron churn mantienen o aumentan su actividad.

### 3. Interacciones con la app
- Churners: menor login, menor tiempo en app, menos clicks en productos financieros.
- No churners: más consistencia en el uso.

### 4. Factores externos
- Podés simular inflación, tasa de interés, etc.
- Definí reglas: clientes de menor salario son más sensibles → más probabilidad de churn cuando inflación sube.

````python
import random

abandono_cust = []

for exited in df_churn_man['Exited']:
    if exited == 1:
        if random.random() < 0.75:
            # 75% de churners en los últimos 6 meses
            rand_idx = int(np.random.triangular(0, len(fechas_abandono)-1, len(fechas_abandono)-1))
            abandono_cust.append(fechas_abandono[rand_idx])
        else:
            # 25 % churners en cualquier fecha del período
            abandono_cust.append(random.choice(fechas))
    else:
        abandono_cust.append(np.nan)
````

````python
# 6) Fechas: asignar fechas según churn/no-churn
# build list of possible dates
start_date = pd.to_datetime('2023-01-01')
end_date   = pd.to_datetime('2025-01-12')
all_dates = pd.date_range(start_date, end_date, freq='D')

# prepare dictionaries for client-level params
client_abandon = df_clients.set_index('CustomerId')['ExitDate'].to_dict()
client_exited  = df_clients.set_index('CustomerId')['Exited'].to_dict()

# exponential decay params
half_life = 30.0
lam = np.log(2)/half_life

# function to sample dates vectorized per client
def sample_dates_for_client(n, cid):
    if client_exited.get(cid,0)==1 and not pd.isna(client_abandon.get(cid)):
        end = pd.to_datetime(client_abandon[cid]) - pd.Timedelta(days=1)
        start = end - pd.Timedelta(days=180)  # window of 6 months
        rng_days = (end - start).days + 1
        if rng_days <= 0:
            # fallback to uniform before end_date
            choices = pd.date_range(start_date, end, freq='D')
            return np.random.choice(choices, size=n)
        # decay probabilities: more weight near end
        days = np.arange(rng_days)
        weights = np.exp(-lam*(rng_days-1 - days))  # higher at recent days
        weights = weights / weights.sum()
        choices = pd.date_range(start, end, freq='D')
        return np.random.choice(choices, size=n, p=weights)
    else:
        # active client: uniform across all_dates
        return np.random.choice(all_dates, size=n)

# vectorized assignment of dates to df_assigned
df_assigned['date'] = None
grouped = df_assigned.groupby('CustomerId').size().reset_index(name='n_tx')
for cid, n in grouped.values:
    dates = sample_dates_for_client(n, cid)
    df_assigned.loc[df_assigned['CustomerId']==cid, 'date'] = dates
````

````python
# --------------------------
# Parámetros ajustables
# --------------------------
monetary_min = 1.0        # monto mínimo posible
monetary_max = 50000.0    # máximo razonable (ajustalo a lo que consideres realista)
clip_low_pct = 0.01       # recortar bottom 1% de amounts
clip_high_pct = 0.99      # recortar top 1% de amounts
eps = 1e-6

# --------------------------
# 0) Preparar clients: obtener EstimatedSalary por CustomerId
#    (si ya tienes max_tx_cap, lo recalculamos de forma robusta)
# --------------------------
clients = df_clients_clean[['CustomerId','EstimatedSalary','Balance']].drop_duplicates().set_index('CustomerId')

# 1) compresión de salario: log1p + min-max
clients['salary_log'] = np.log1p(clients['EstimatedSalary'].fillna(0.0))
scaler = MinMaxScaler()
clients['salary_norm'] = scaler.fit_transform(clients[['salary_log']])

# 2) mapear a rango monetario (cap por cliente)
clients['max_tx_cap'] = clients['salary_norm'] * (monetary_max - monetary_min) + monetary_min

# si querés dar algo de peso al balance, podrías usar:
# clients['max_tx_cap'] = 0.85*clients['max_tx_cap'] + 0.15*(clients['Balance'].clip(upper=monetary_max))

# 3) merge con df (si df ya contiene CustomerId)
df = df_tx_clean.merge(clients[['max_tx_cap']], left_on='CustomerId', right_index=True, how='left')

# 4) recortar amounts por percentiles para evitar outliers extremos
q_low, q_high = df['amount'].quantile([clip_low_pct, clip_high_pct])
df['amount_clipped'] = df['amount'].clip(lower=q_low, upper=q_high)

# 5) normalizar clipped amount a rango 0..1 (si q_high==q_low proteger)
den = (q_high - q_low) if (q_high - q_low) > 0 else eps
df['amount_ratio'] = (df['amount_clipped'] - q_low) / den
df['amount_ratio'] = df['amount_ratio'].clip(0,1)

# 6) calcular amount_scaled
df['amount_scaled'] = (df['amount_ratio'] * df['max_tx_cap']).clip(lower=monetary_min).round(2)
````

````python
# -------------------------
# 3) RELACIÓN CLIENTE - TRANSACCIÓN
# -------------------------

tx_per_customer = df_tx_scaled.groupby('CustomerId')['amount_scaled'].agg(['count','mean','median','max'])
print(tx_per_customer.describe())

# Distribución número de transacciones por cliente
plt.figure(figsize=(8,4))
sns.histplot(tx_per_customer['count'], bins=50, log_scale=(False, True))
plt.title("Número de transacciones por cliente (escala y log)")
plt.savefig(os.path.join(IMG_PATH, "df_clients_num_transacciones.png"), dpi=300)
plt.show()
````

````python
# -------------------------
# 4) CHURN
# -------------------------

# Distribución de churn
print(df_clients_clean['Exited'].value_counts(normalize=True))

# Media y mediana de transacciones por churn
tx_per_customer = tx_per_customer.reset_index().merge(df_clients_clean[['CustomerId','Exited']], on='CustomerId', how='left')
print(tx_per_customer.groupby('Exited')[['count','mean','median']].mean())
````

````python
# -------------------------
# 5) CORRELACIONES
# -------------------------

corr_columns = numeric_cols_clients
corr_columns.append('Exited')
plt.figure(figsize=(10,8))
sns.heatmap(df_clients_clean[corr_columns].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlación variables clientes")
plt.savefig(os.path.join(IMG_PATH, "corr_matrix.png"), dpi=300)
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x='EstimatedSalary', y='amount_scaled', data=df_clients_clean.merge(df_tx_scaled, on='CustomerId', how='right').sample(10000))
plt.title("Salary vs amount_scaled (sample 10k)")
plt.savefig(os.path.join(IMG_PATH, "scatter_salarios_transacciones.png"), dpi=300)
plt.show()
````
