## <font color='#315aa0'>**Mapa a seguir**</font>

### 1. Preparaci√≥n de clientes (df_churn)

* **Mantener todas las columnas originales de df_churn, especialmente:**

    - CustomerId (ID √∫nico)
    - Exited (target)
    - Balance, Tenure, CreditScore, Age, NumOfProducts, etc.
    - Generar columna ‚ÄúMes de abandono‚Äù para clientes que hicieron churn (Exited=1):
        - Aleatorio entre 0 y 6 meses antes del final del per√≠odo simulado.
        - Esto definir√° hasta cu√°ndo tendr√°n transacciones.
    - Crear columna ‚ÄúBalance inicial‚Äù que ser√° el punto de partida para generar oldbalanceOrig y newbalanceOrig en transacciones.

### 2. Preparaci√≥n de transacciones (df_paysim)

* **Filtrar y seleccionar columnas relevantes:**

    - step ‚Üí ser√° transformado en fecha
    - type ‚Üí tipo de transacci√≥n
    - amount ‚Üí monto, escalado y limitado por balance
    - nameOrig ‚Üí ser√° reemplazado por CustomerId
    - Opcional: oldbalanceOrg, newbalanceOrig ‚Üí recalculados para coherencia
    - isFraud ‚Üí puede generar feature de ‚Äúfraudulent activity‚Äù

* **Escalar montos:**

    - `amount_scaled = (amount_original / max_amount_paysim) * balance_cliente`


Esto asegura que ninguna transacci√≥n supere el balance del cliente.

* **Convertir step en fechas:**

    - Definir fecha inicial ficticia, p.ej. "2023-01-01".
    - Transformar step en d√≠as o meses:

        - `fecha_transaccion = fecha_inicio + timedelta(days=step)`

Permite agregar features temporales: actividad por mes, √∫ltimos 6 meses, etc.

### 3. Asignaci√≥n de transacciones a clientes

* Reemplazar nameOrig por CustomerId de df_churn.

* **Reglas de asignaci√≥n:**

    - Clientes que no hicieron churn (Exited=0) ‚Üí asignar transacciones a lo largo de todo el per√≠odo simulado.
    - Clientes que hicieron churn (Exited=1) ‚Üí asignar transacciones hasta ‚Äúmes de abandono‚Äù.

* **Opciones de distribuci√≥n:**
    - Aleatoria, pero proporcional al balance o al tenure.
    - Para que haya variedad en la cantidad de transacciones por cliente.

### 4. Generaci√≥n de balances coherentes

**Para cada cliente:**

oldbalanceOrg = balance_anterior
newbalanceOrig = balance_anterior - amount_scaled (para retiros) o + amount_scaled (para dep√≥sitos)

Esto mantiene consistencia interna en la historia de transacciones.

### 5. Agregaci√≥n y creaci√≥n de features por cliente

Estas ser√°n las columnas que usar√≠a el modelo de ML:

**Activity Features:**

* Total de transacciones
* Promedio de monto por transacci√≥n
* N√∫mero de transacciones por tipo (CASH_OUT, PAYMENT, etc.)
* √öltimo mes de actividad
* Varianza de montos

**Balance Features:**

* Promedio de saldo (newbalanceOrig)

* √öltimo saldo registrado

**Fraud / Riesgo:**

* N√∫mero de transacciones fraudulentas (isFraud)
* Proporci√≥n de transacciones grandes (> X% del saldo)

**Temporal:**

* Tendencia de actividad (disminuci√≥n antes del churn)
* Mes desde la √∫ltima transacci√≥n

### 6. Dataset final para ML

## üîë Principios para evitar ‚Äúdataset basura‚Äù

### 1. Defin√≠ hip√≥tesis de negocio realistas
Antes de tocar los datos, escrib√≠ supuestos que tengan sentido en un banco real:

- Clientes con menos interacciones en la app ‚Üí m√°s probabilidad de churn.
- Disminuci√≥n en frecuencia de transacciones en los √∫ltimos meses ‚Üí mayor riesgo de churn.
- Edad y salario influyen en la estabilidad del cliente (j√≥venes y bajos salarios = m√°s churn).
- Factores externos como inflaci√≥n alta ‚Üí m√°s churn en clientes de bajo ingreso.

> üëâ Esto asegura que los patrones que generes no sean aleatorios, sino consistentes con lo que podr√≠a pasar en la realidad.

---

### 2. Us√° correlaciones en lugar de ruido puro
En lugar de asignar variables externas o de interacci√≥n al azar:

- Haz que est√©n correlacionadas con churn de forma controlada.

**Ejemplo:**
- Clientes que hicieron churn tienen un 40% menos de interacciones con la app en los √∫ltimos 3 meses.
- Clientes de churn tienen un 20% m√°s de transacciones fallidas o rechazadas.
- Factores externos (ej: inflaci√≥n simulada) afectan m√°s a clientes con Balance < X.

> üëâ Esto genera se√±ales diferenciales para el modelo.

---

### 3. Introduce variabilidad dentro de cada grupo
No todos los clientes que hicieron churn deben tener baja actividad. Introduc√≠ distribuciones:

- 60% de churners tienen muy baja interacci√≥n.
- 30% tienen actividad media.
- 10% siguen siendo activos hasta el final.

> üëâ Esto evita que el modelo aprenda reglas triviales (del tipo: ‚Äúsi transacciones = 0 ‚Üí churn‚Äù).

---

### 4. Us√° ruido controlado (stochastic noise)
Un poco de azar es necesario para que no todo sea determin√≠stico, pero debe ser peque√±o en comparaci√≥n con la se√±al principal.

**Ejemplo:**  
N√∫mero de interacciones promedio de un churner = 20, con desviaci√≥n est√°ndar de ¬±5.

---

### 5. Validaci√≥n manual antes del modelo
Antes de entrenar, explor√° el dataset final con gr√°ficos:

- Histograma de transacciones churn vs no churn.
- Interacciones app churn vs no churn.
- Factores externos seg√∫n balance.

> üëâ Si ves diferencias razonables entre grupos, entonces tu dataset tiene patrones √∫tiles.

---

## üîß Concretando para tus 4 datasets

### 1. Churn (base)
- Target + atributos est√°ticos.

### 2. Transacciones
- Clientes que hicieron churn deben mostrar una ca√≠da de actividad en los √∫ltimos meses.
- Clientes que no hicieron churn mantienen o aumentan su actividad.

### 3. Interacciones con la app
- Churners: menor login, menor tiempo en app, menos clicks en productos financieros.
- No churners: m√°s consistencia en el uso.

### 4. Factores externos
- Pod√©s simular inflaci√≥n, tasa de inter√©s, etc.
- Defin√≠ reglas: clientes de menor salario son m√°s sensibles ‚Üí m√°s probabilidad de churn cuando inflaci√≥n sube.

````python
# --------------------------
# Par√°metros ajustables
# --------------------------
monetary_min = 1.0        # monto m√≠nimo posible
monetary_max = 50000.0    # m√°ximo razonable (ajustalo a lo que consideres realista)
clip_low_pct = 0.01       # recortar bottom 1% de amounts
clip_high_pct = 0.99      # recortar top 1% de amounts
eps = 1e-6

# --------------------------
# 0) Preparar clients: obtener EstimatedSalary por CustomerId
#    (si ya tienes max_tx_cap, lo recalculamos de forma robusta)
# --------------------------
clients = df_clients_clean[['CustomerId','EstimatedSalary','Balance']].drop_duplicates().set_index('CustomerId')

# 1) compresi√≥n de salario: log1p + min-max
clients['salary_log'] = np.log1p(clients['EstimatedSalary'].fillna(0.0))
scaler = MinMaxScaler()
clients['salary_norm'] = scaler.fit_transform(clients[['salary_log']])

# 2) mapear a rango monetario (cap por cliente)
clients['max_tx_cap'] = clients['salary_norm'] * (monetary_max - monetary_min) + monetary_min

# si quer√©s dar algo de peso al balance, podr√≠as usar:
# clients['max_tx_cap'] = 0.85*clients['max_tx_cap'] + 0.15*(clients['Balance'].clip(upper=monetary_max))

# 3) merge con df (si df ya contiene CustomerId)
df = df_tx_clean.merge(clients[['max_tx_cap']], left_on='CustomerId', right_index=True, how='left')

# 4) recortar amounts por percentiles para evitar outliers extremos
q_low, q_high = df['amount'].quantile([clip_low_pct, clip_high_pct])
df['amount_clipped'] = df['amount'].clip(lower=q_low, upper=q_high)

# 5) normalizar clipped amount a rango 0..1 (si q_high==q_low proteger)
den = (q_high - q_low) if (q_high - q_low) > 0 else eps
df['amount_ratio'] = (df['amount_clipped'] - q_low) / den
df['amount_ratio'] = df['amount_ratio'].clip(0,1)

# 6) calcular amount_scaled
df['amount_scaled'] = (df['amount_ratio'] * df['max_tx_cap']).clip(lower=monetary_min).round(2)
````
